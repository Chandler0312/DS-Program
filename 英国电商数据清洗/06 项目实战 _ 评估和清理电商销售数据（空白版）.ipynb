# 项目：评估和清理英国电商公司销售数据

## 分析目标

此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。

本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

## 简介

原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。

数据每列的含义如下：
- `InvoiceNo`: 发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。
- `StockCode`: 产品代码。5位数，作为产品的唯一标识符。
- `Description`: 产品名称。
- `Quantity`: 产品在交易中的数量。
- `InvoiceDate`: 发票日期和时间。交易发生的日期和时间。
- `UnitPrice`: 单价。价格单位为英镑（£）。
- `CustomerID`: 客户编号。5位数，作为客户的唯一标识符。
- `Country`: 国家名称。客户所居住的国家的名称。

## 读取数据

导入数据分析所需要的库，平通过pandas的`read_csv`函数，将原数据文件"e_commerce.csv"里的数据内容，解析为DataFrame，并赋值给变量`df`

```python
import pandas as pd
df = pd.read_csv("e_commerce.csv")
df.head()
```

## 评估数据

#### **评估数据整洁度**

```python
df.sample(10)
```

结论：抽样10行结果来看，符合“每列是一个变量，每行是一个观察值，每个单元格十一个值”，具体看来每行是关于某商品的一次交易，每列是交易相关变量，因此不存在结构性问题

#### **评估数据干净程度**

```python
df.info()
```

结论：通过输出结果看来，数据总共有541909条观察值，而 ``Description（540455）, CustomerID（406829）`` 这两个变量存在缺失值。  
     另外，``InvoiceDate`` 变量的数据类型应该为日期，而`CustomerID`变量无需统计运算，数据类型应该为字符串，应该进行数据格式转换。

#### **评估缺失数据**

根据条件提取出`Description`列中的缺失值

```python
df[df['Description'].isnull()]
```

有1454条交易数据缺失Description变量值。  
从输出结果看来，缺失`Description`的交易数据，`UnitPrice`都为0。为了验证猜想我们增加筛选条件，看是否存在`Description`变量缺失且`UnitPrice`不为0的数据

```python
df[(df['Description'].isnull()) & (df['UnitPrice'] != 0)]
```

筛选出来结果为0条，说明假设成立，缺失`Description`值的数据，同样也不具备有效的`UnitPrice`值。
`Description`与``UnitPrice``都是商品分析的重要变量，如果他们同时缺失/无效，我们认为这些数据无法提供有效意义，因此这些数据可以在后续进行删除。

根据条件提取出`CustomerID`列中的缺失值

```python
df[df['CustomerID'].isnull()]
```

`CustomerID`表示客户编号，不受分析畅销产品的必要变量。并且从结果来看，有些`CustomerID`缺失的交易数据仍然有小，因此保留此变量为空的观察值。

#### **评估重复数据**

根据数据变量含义看来，虽然`InvoiceNO`,`StockCode`,`CustomerID`都是唯一标识符但是：  
一次交易可能包含多件商品，因此`InvoiceNO`可以存在重复  
不同交易可以包含同件商品，因此`StockCode`可以存在重复  
顾客可以进行多次交易或下单多个商品，因此`CustomerID`可以存在重复

#### **评估不一致数据**

不一致数据可能存在于`Country`变量中，我们可以查看是否存在多个不同值知道同一国家的情况

```python
df['Country'].value_counts()
```

从筛选结果看来，`USA`,`United States`均在表示美国，`United Kingdom`,`UK`,`U.K.`均在表示英国，因此需要在数据清洗阶段对这些值进行统一，只保留一个指代值。

#### **评估无效或错误数据**

通过`describe`方法，对数值统计信息快速了解

```python
df.describe(include = 'all')
```

```python
df[df['Quantity'] < 0]
```

从筛选结果来看，`Quantity`变量为复数的观察值，`InvoiceNo`似乎均以C开头，表示订单被取消。  
但此猜测不一定准确，为了验证猜想，我们增加筛选条件，看是否存在`Quantity`变量为负数且`InvoiceNo`不已C开头的观察值

```python
df[(df['Quantity'] < 0) & (df['InvoiceNo'].str[0] != 'C')]
```

以上猜想被证实错误，因为还存在1336行数据`Quantity`变量为复数且`InvoiceNo`不以C开头  
但根据以上输出结果，这些筛选出的观察值的`UnitPrice`观察值均为0，因此增加`UnitPrice`的条件验证

```python
df[(df['Quantity'] < 0) & (df['InvoiceNo'].str[0] != 'C') & (df['UnitPrice'] != 0)]
```

根据输出结果，猜想得到验证，当Quantity 变量为负数的时候，观蔡值满足以下条件之一  
- `Invoiceso` 以C开头，表示订单被取消  
- `Uniterice` 为0，说明单价为O英镑

这些交易数据均不是有效成交数据，不贡献销售，不在后续分析范四内，因此我们将在数据清理步骤，将`Quantity`为负数的观察值删除。

接下来筛选`UnitPrice`为复数的观察值。

```python
df[df['UnitPrice'] < 0]
```

从输出结果来看，`UnitPrice`为负数的观察值都是坏账调账，不属于实际商品交易数据，因此也在数据清洗阶段将其删除。

## 清理数据

根据前面评估部分得到的结论，清洗阶段步骤包括：  
- 把`InvoiceDate`变量数据类型转换为日期格式
- 把`CustomerID`变量数据类型转化为字符串
- 把`Description`变量缺失的观察值删除
- 把`Country`变量值`USA`转化为`United States`, `UK`,`U.K.`转化为`United Kingdom`
- 把`Quantity`变量值为负数的观察值删除
- 把`UnitPrice`变量值为负数的观察值删除

为区分开经过清理的数据和原始数据，我们创建新的变量`cleaned_data`，让它成为`df`的副本。之后的清洗步骤都被使用在`cleaned_data`上

```python
cleaned_data = df.copy()
```

把`InvoiceDate`变量数据类型转换为日期格式

```python
cleaned_data['InvoiceDate'] = pd.to_datetime(cleaned_data['InvoiceDate'])
cleaned_data['InvoiceDate']
```

把`CustomerID`变量数据类型转化为字符串

```python
cleaned_data['CustomerID'] = cleaned_data['CustomerID'].astype(str)
cleaned_data['CustomerID']
```

把`CustomerID`变量值结尾的".0"删除

```python
cleaned_data['CustomerID'] = cleaned_data['CustomerID'].str.slice(0,-2)
cleaned_data['CustomerID']
```

把`Description`变量缺失的观察值删除

```python
cleaned_data.dropna(subset = ["Description"], axis = 0, inplace = True)
```

把`Country`变量值`USA`转化为`United States`, `UK`,`U.K.`转化为`United Kingdom`

```python
cleaned_data['Country'].replace({'USA' : 'United States'}, inplace = True)
cleaned_data['Country'].replace(['UK','U.K.'], 'United Kingdom', inplace = True)
```

把`Quantity`变量值为负数的观察值删除

```python
cleaned_data = cleaned_data[cleaned_data['Quantity'] >= 0]
```

把`UnitPrice`变量值为负数的观察值删除

```python
cleaned_data = cleaned_data[cleaned_data['UnitPrice'] >= 0]
```

观察清洗后数据并且采样30条

```python
cleaned_data.sample(30)
```

## 保存清理后的数据

完成数据清洗，把干净数据保存在新的文件里，文件名为`'e_commerce_cleaned.csv'`

```python
cleaned_data.to_csv('e_commerce_cleaned.csv', index = False)
```
